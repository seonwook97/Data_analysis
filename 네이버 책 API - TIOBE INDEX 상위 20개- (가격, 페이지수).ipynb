```python
def gen_search_url(search_text, start_num, disp_num):
    base = "https://openapi.naver.com/v1/search"
    node = "/book_adv.json" 
    param_title = "?d_titl=" + urllib.parse.quote(search_text)
    param_start = "&start=" + str(start_num)
    param_disp = "&display=" + str(disp_num)
    param_catg = "&d_catg=280" # 컴퓨터/IT 카테고리 넘버 = 280
    
    return base + node + param_title + param_start + param_disp + param_catg 
```

```python
def get_result_onpage(url):
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id", client_id)
    request.add_header("X-Naver-Client-Secret", client_secret)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    print("[%s] Url Request Success" % datetime.datetime.now())
    print(response, response.getcode(), response.code, response.status) 
    return json.loads(response.read().decode("utf-8"))
```

```python
def delete_tag(input_str):
    input_str = input_str.replace("<b>", "")
    input_str = input_str.replace("</b>", "")
    return input_str 
```

```python
def get_fields(json_data):
    title = [delete_tag(each["title"]) for each in json_data["items"]]
    link = [each["link"] for each in json_data["items"]]
    author = [each["author"] for each in json_data["items"]]
    price = [each["price"] for each in json_data["items"]]
    discount = [each["discount"] for each in json_data["items"]]
    publisher = [each["publisher"] for each in json_data["items"]]
    isbn = [each["isbn"] for each in json_data["items"]]
    pubdate = [each["pubdate"] for each in json_data["items"]]
    result_pd = pd.DataFrame({
        "제목": title, 
        "링크": link, 
        "저자": author, 
        "가격": price,
        "할인 가격": discount,
        "isbn": isbn,
        "출판사": publisher,
        '출간일': pubdate
        
    }, columns=["제목", "저자", "출판사", "가격", "할인 가격", "출간일", "isbn", "링크"])
    return result_pd
```

```python
langs = ['Assembly language', 'Classic Visual Basic', 'Go', 'MATLAB', '매트랩', 'Groovy', '그루비', 'Ruby', '루비', 'Swift', '스위프트', 'Fortran', '포트란', 'Perl', '펄', 'Delphi', '델파이']

def search_func(name):
    result_lang = []
    for n in range(1, 1000, 100):
        url = gen_search_url(name, n, 100)
        json_result = get_result_onpage(url)
        pd_result = get_fields(json_result)
        pd_result['검색명'] = '{}'.format(name)
        result_lang.append(pd_result)
    result_lang = pd.concat(result_lang)
    return result_lang
```

```python
langs_list = [] 

for i in range(len(langs)):
    langs_list.append(search_func(langs[i]))
```
