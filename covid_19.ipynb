```python
dates = country_confirmed.keys()
world_cases = []
total_deaths = [] 
mortality_rate = []

for i in dates:
    confirmed_sum = country_confirmed[i].sum()
    death_sum = country_deaths[i].sum()    
    
    world_cases.append(confirmed_sum)
    total_deaths.append(death_sum)
    
    mortality_rate.append(death_sum/confirmed_sum)
```

```python
def daily_increase(data):
    d = [] 
    for i in range(len(data)):
        if i == 0:
            d.append(data[0])
        else:
            d.append(data[i]-data[i-1])
    return d 

def moving_average(data, window_size):
    moving_average = []
    for i in range(len(data)):
        if i + window_size < len(data):
            moving_average.append(np.mean(data[i:i+window_size]))
        else:
            moving_average.append(np.mean(data[i:len(data)]))
    return moving_average

# window size
window = 7

# confirmed cases
world_daily_increase = daily_increase(world_cases)
world_confirmed_avg= moving_average(world_cases, window)
world_daily_increase_avg = moving_average(world_daily_increase, window)

# deaths
world_daily_death = daily_increase(total_deaths)
world_death_avg = moving_average(total_deaths, window)
world_daily_death_avg = moving_average(world_daily_death, window)
```

```python
def country_plot(x, y1, y2, y3, country):
    # window is set as 14 in in the beginning of the notebook 
    confirmed_avg = moving_average(y1, window)
    confirmed_increase_avg = moving_average(y2, window)
    death_increase_avg = moving_average(y3, window)
#     recovery_increase_avg = moving_average(y4, window)
    
    plt.figure(figsize=(16, 10))
    plt.plot(x, y1)
    plt.plot(x, confirmed_avg, color='red', linestyle='dashed')
    plt.legend(['{} 일일 누적 확진자 수'.format(country), '주 평균 누적 확진자 수'], prop={'size': 15})
    plt.title('{} 일별 누적 코로나 확진자 수'.format(country), size=25)
    plt.xlabel('2020년 1월 22일 부터', size=20)
    plt.ylabel('확진자수', size=20)
    plt.xticks(size=15)
    plt.yticks(size=15)
    plt.show()

    plt.figure(figsize=(16, 10))
    plt.bar(x, y2)
    plt.plot(x, confirmed_increase_avg, color='red', linestyle='dashed')
    plt.legend(['주 평균 확진자 수', '{} 일일 확진자 수'.format(country)], prop={'size': 15})
    plt.title('{} 일별 코로나 확진자 수'.format(country), size=25)
    plt.xlabel('2020년 1월 22일 부터', size=20)
    plt.ylabel('확진자수', size=20)
    plt.xticks(size=15)
    plt.yticks(size=15)
    plt.show()

    plt.figure(figsize=(16, 10))
    plt.bar(x, y3)
    plt.plot(x, death_increase_avg, color='red', linestyle='dashed')
    plt.legend(['주 평균 사망자 수', '{} 일일 사망자 수'.format(country)], prop={'size': 15})
    plt.title('{} 일별 코로나 사망자 수'.format(country), size=25)
    plt.xlabel('2020년 1월 22일 부터', size=20)
    plt.ylabel('사망자수', size=20)
    plt.xticks(size=15)
    plt.yticks(size=15)
    plt.show()

#     plt.figure(figsize=(16, 10))
#     plt.bar(x, y4)
#     plt.plot(x, recovery_increase_avg, color='red', linestyle='dashed')
#     plt.legend(['Moving Average {} Days'.format(window), '{} Daily Increase in Confirmed Recoveries'.format(country)], prop={'size': 20})
#     plt.title('{} Daily Increases in Recoveries'.format(country), size=30)
#     plt.xlabel('Days Since 1/22/2020', size=30)
#     plt.ylabel('# of Cases', size=30)
#     plt.xticks(size=20)
#     plt.yticks(size=20)
#     plt.show()
      
# helper function for getting country's cases, deaths, and recoveries        
def get_country_info(country_name):
    country_cases = []
    country_deaths = []
#     country_recoveries = []  
    
    for i in dates:
        country_cases.append(confirmed_df[confirmed_df['Country/Region']==country_name][i].sum())
        country_deaths.append(deaths_df[deaths_df['Country/Region']==country_name][i].sum())
#         country_recoveries.append(recoveries_df[recoveries_df['Country/Region']==country_name][i].sum())
    return (country_cases, country_deaths)
    
    
def country_visualizations(country_name):
    country_info = get_country_info(country_name)
    country_cases = country_info[0]
    country_deaths = country_info[1]
    
    country_daily_increase = daily_increase(country_cases)
    country_daily_death = daily_increase(country_deaths)
#     country_daily_recovery = daily_increase(country_recoveries)
    
    country_plot(adjusted_dates, country_cases, country_daily_increase, country_daily_death, country_name)
```

```python
avg_rate = [np.mean(Korea_daily_df['일일 확진자 대비 사망자 비율']), np.mean(Japan_daily_df['일일 확진자 대비 사망자 비율']), np.mean(USA_daily_df['일일 확진자 대비 사망자 비율']), np.mean(UK_daily_df['일일 확진자 대비 사망자 비율']), np.mean(Ger_daily_df['일일 확진자 대비 사망자 비율'])]
avg_death_rate_per_confirmed = pd.DataFrame({'국가':['한국', '일본', '미국', '영국', '독일'],
                                             '일일 확진자 대비 사망자 비율 평균': avg_rate})
avg_death_rate_per_confirmed.set_index('국가', inplace=True)
avg_death_rate_per_confirmed
avg_death_rate_per_confirmed.sort_values(by='일일 확진자 대비 사망자 비율 평균').plot(
    kind='barh', 
    figsize=(10, 10),
    title='국가별 일별 확진자 대비 사망자 비율 평균 (%)',
    xlabel='국가',
    color='purple')
```

- 독일의 값은 코딩과정에서 문제가 있었는지 수정을 해도 inf가 반환되어 일단 결과 해석에서는 잠시 제외하였다.

# 문제 (4)

- 문제3의 경우에서 다시 인구수별 비율로 표현하려는 노력도 함께 해주세요.

- 5개 국가뿐이니 인구수는 단순 검색을 통해 용이하게 구할 수 있을 것으로 보입니다.

```python
# 국가별 현재까지 누적 확진자, 사망자 df 생성
confirmed_pop_rate = pd.DataFrame({'현재 누적 사망자 수': country_deaths.iloc[0:5, 629],
                                   '현재 누적 확진자 수': country_confirmed.iloc[0:5, 629]})

# 컬럼 순서 변경
confirmed_pop_rate = confirmed_pop_rate[['현재 누적 확진자 수', '현재 누적 사망자 수']]

# 국가별 인구수 컬럼 추가
confirmed_pop_rate['인구수'] = [51611400, 125960000, 332915074, 67886011, 83783942 ]

# 인구 대비 누적 확진자, 사망자 컬럼 추가
confirmed_pop_rate['인구 대비 누적 확진자 비율'] = confirmed_pop_rate['현재 누적 확진자 수'] / confirmed_pop_rate['인구수'] * 100
confirmed_pop_rate['인구 대비 누적 사망자 비율'] = confirmed_pop_rate['현재 누적 사망자 수'] / confirmed_pop_rate['인구수'] * 100

confirmed_pop_rate
```

```python
confirmed_pop_rate['인구 대비 누적 확진자 비율'].sort_values().plot(
    kind='barh', 
    figsize=(10, 10),
    title='국가별 인구 대비 코로나 누적 확진자 비율 (%)',
    xlabel='국가',
    color='darkblue')
```

```python
confirmed_pop_rate['인구 대비 누적 사망자 비율'].sort_values().plot(
    kind='barh', 
    figsize=(10, 10),
    title='국가별 인구 대비 코로나 누적 사망자 비율 (%)',
    xlabel='국가',
    color='darkred')
```

- 국가별 인구 대비 코로나 누적 확진자 비율은 미국이 약 13%로 가장 높았고, 한국이 약 0.6%로 가장 낮았다
- 국가별 인구 대비 코로나 누적 사망자 비율 역시 미국이 약 0.2%로 가장 높았으며, 한국이 약 0.005%로 가장 낮았다

# 문제 (5)

- 전체 데이터의 시계열 분석을 통해 전체 트렌드와 주기별 특성을 관찰해 주세요. 어떤 특이점이 있나요?
- 5개 각 국가의 데이터에서는 동일과정에서 어떤 특이한 현상이 있나요? 

```python
ds = pd.date_range('2020-01-22', periods=630, freq='D')

df_confirmed = pd.DataFrame({'ds':ds, 
                             '한국 누적 확진자 수': country_confirmed.loc['한국'],
                             '일본 누적 확진자 수': country_confirmed.loc['일본'],
                             '미국 누적 확진자 수': country_confirmed.loc['미국'],
                             '영국 누적 확진자 수': country_confirmed.loc['영국'],
                             '독일 누적 확진자 수': country_confirmed.loc['독일'],})

df_deaths = pd.DataFrame({'ds':ds, 
                          '한국 누적 사망자 수': country_deaths.loc['한국'],
                          '일본 누적 사망자 수': country_deaths.loc['일본'],
                          '미국 누적 사망자 수': country_deaths.loc['미국'],
                          '영국 누적 사망자 수': country_deaths.loc['영국'],
                          '독일 누적 사망자 수': country_deaths.loc['독일'],})
```

```python
# 데이터형 변환 object => int
country = ['한국', '일본', '미국', '영국', '독일']

for a in country:
    df_confirmed[a + ' ' + '누적 확진자 수'] = df_confirmed[a + ' ' + '누적 확진자 수'].astype('int')
    df_deaths[a + ' ' + '누적 사망자 수'] = df_deaths[a + ' ' + '누적 사망자 수'].astype('int')
```

```python
from fbprophet import Prophet
```

```python
# 한국 누적 확진자 수 Forecast(90일)
subdf = df_confirmed[['ds', '한국 누적 확진자 수']].dropna()
subdf = subdf.rename(columns={'ds':'ds', '한국 누적 확진자 수':'y'})

m = Prophet(yearly_seasonality=True, daily_seasonality=True)
m.fit(subdf);

future = m.make_future_dataframe(periods=90)
forecast = m.predict(future)

m.plot(forecast);
```
